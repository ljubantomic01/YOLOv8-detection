# Skripta za testiranje modela detekcije
Ova skripta služi kako bi se testirao dotrenirani yolov8n model, koji detektira ovce i sove.

## Koraci za detekciju:

**1.** Preuzeti dijeljenu mapu 'yolov8_detection' i pronaći datoteku 'best.pt' koji se nalazi u putanji direktorija: 'yolov8_detection\runs\detect\train\weights\', ili samo preuzeti tu datoteku na GitHubu : https://github.com/ljubantomic01/YOLOv8-detection.git;

**2.** Uploadati navedeni file u Google Colab okruženje, u sektor Datoteke;

**3.** Pokrenuti ćelije s kodom, po redu, najbolje jednu po jednu.

  **3.1.** Prilikom detekcije (ćelija koda br. 3) potrebno je paziti da se napiše ispravno ime modela (u ovome slučaju best.pt) i da se da ispravna putanja (eng. *source*) do slike ili direktorija u kojem su slike, relativno na /content, koji je primarni direktorij u Google Colabu, koji sadrži ostale, koji služe pri korištenju modela (direktorij 'sample_data' je default direktorij). Isto tako, rezultati se spremaju u 'runs/predict/' koji se automatski prvi put generira.

  **3.2.** U ćeliji koda 4, bitno je paziti, ako se model više puta pokreće, da se broj direktorija povećava (predict, predict2, predict3,...), pa je važno dodati broj kako bi se rezultati ispravno prikazali.
